```python
import pickle 

import scanpy as sc
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
geneset_path = "/path/to/genesets/"
data_path = "/path/to/data/"

classifier_save_path = '/path/to/save/celltype/classifier/'
```

```python
adata = sc.read_h5ad(f"{data_path}20240229_from_rocio_for_manuscript.h5ad")
adata.X = adata.layers['logcounts']
with open(f'{ge}functional_genes.txt', "r") as f:
    genes = f.read().splitlines()
```

## CD4 and CD8

```python
sel_adata = adata[:, genes]
```

```python
X = np.array(sel_adata.X.todense())
y = sel_adata.obs["ident"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, stratify=y)
```

```python
lm = LogisticRegression()
svm = SVC()
rf = RandomForestClassifier()
nn = MLPClassifier()
```

```python
lm.fit(X_train, y_train)
svm.fit(X_train, y_train)
rf.fit(X_train, y_train)
nn.fit(X_train, y_train)
```

```python
print("LM: ", lm.score(X_test, y_test))
print("SVM: ", svm.score(X_test, y_test))
print("RF: ", rf.score(X_test, y_test))
print("NN: ", nn.score(X_test, y_test))
```

```python
svm.fit(X, y)
```

```python
with open(f"{classifier_save_path}20240229_from_rocio_for_manuscript_functional_genes_svm.pkl", "wb") as f:
    pickle.dump(svm, f)
```

```python
CD4_adata = sc.read_h5ad(f"{data_path}CD4.h5ad")
CD8_adata = sc.read_h5ad(f"{data_path}CD8.h5ad")
```

```python
with open(f"{classifier_save_path}CD4_functional_genes_svm.pkl", "wb") as f:
    pickle.dump(CD4_svm, f)
with open(f"{classifier_save_path}CD8_functional_genes_svm.pkl", "wb") as f:
    pickle.dump(CD8_svm, f)
```
